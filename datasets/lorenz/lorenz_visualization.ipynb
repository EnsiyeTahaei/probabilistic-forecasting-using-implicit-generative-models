```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
with open("./lorenz_dataset.pickle", "rb") as infile:
    dataset = pickle.load(infile)
# print(dataset)
```

```python
plt.figure(figsize=(10,10))
colors_list = ["#eeacda", "#ea9192", "#ba9ed5", "#38a638", "#996a60"]
for i in range(5):
    for ts in dataset["x_train"]["condition_{}".format(i)][:1]:
        plt.plot(ts, color=colors_list[i], label="$y_{}$".format(i), alpha = 0.7)
for i in range(4,-1,-1):
    for ts in dataset["x_train"]["condition_{}".format(i)][1:50]:
        plt.plot(ts, color=colors_list[i], alpha = 0.2)
plt.title("Sample condtional window from Lorenz dataset")
plt.xlabel("t [s]")
plt.ylabel("x")
# plt.legend()
leg = plt.legend()
for line in leg.get_lines():
    line.set_linewidth(2)
plt.show()
```

```python
plt.figure(figsize=(12,8))
colors_list = ["#80c680", "#e67d7e", "#bfa4d7", "#ba9a93", "#eeadda"]
for i in range(5):
    plt.hist(dataset["y_train"]["condition_{}".format(i)], bins=80, density=True, range=(-11, 11),edgecolor="r", 
             color=colors_list[i], label="$y_{}$".format(i), alpha=0.6)
plt.title("The distribution of $x_{t+1}$ distinguished by initial values in train set")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.hist(np.concatenate(list(dataset["y_train"].values())), bins=80, density=True, range=(-11, 11),color="coral",edgecolor="r", 
        label="ground truth")
plt.title("Full probability distribution of $x_{t+1}$ in train set")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))
colors_list = ["#80c680", "#e67d7e", "#bfa4d7", "#ba9a93", "#eeadda"]
for i in range(5):
    plt.hist(dataset["y_test"]["condition_{}".format(i)], bins=80, density=True, range=(-11, 11),edgecolor="r", 
             color=colors_list[i], label="$y_{}$".format(i), alpha=0.6)
plt.title("The distribution of $x_{t+1}$ distinguished by initial values in train set")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.legend()
plt.show()
```
